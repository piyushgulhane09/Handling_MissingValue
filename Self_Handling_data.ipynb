## Handling Null Values with (MMM) Mean, Median, Mode

```python
np.mean(df['age'])
```

```python
df['age']=df['age'].fillna(np.mean(df['age']))
sns.heatmap(df.isnull())
```

```python
mode = list(df.deck.value_counts().index)[0]
df['deck']=df['deck'].fillna(mode)
```
